```python
import os

data_dir = "../data/chest_xray"
print(os.listdir(data_dir))
print(os.listdir(os.path.join(data_dir, "train")))
```

```python
import matplotlib.pyplot as plt
import cv2
import random
import os

train_normal_dir = "../data/chest_xray/train/NORMAL"
train_pneumonia_dir = "../data/chest_xray/train/PNEUMONIA"

# choose random file
img_path = os.path.join(train_normal_dir, random.choice(os.listdir(train_normal_dir)))

# read and show
img = cv2.imread(img_path)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.imshow(img)
plt.title("Sample X-ray (Normal)")
plt.axis("off")
```

```python
img_path = os.path.join(train_pneumonia_dir, random.choice(os.listdir(train_pneumonia_dir)))

img = cv2.imread(img_path)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.imshow(img)
plt.title("Sample X-ray (Pneumonia)")
plt.axis("off")
```
